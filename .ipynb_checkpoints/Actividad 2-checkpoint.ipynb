### 3. Programe el método de sustitución regresiva. Analice el número de operaciones del algoritmo para matrices de tamaño 3 × 3.
#### Método de sustitución regresiva:
#### $$x_i=\frac{b_i-\sum\limits_{j=i+1}^{n}{a_{i,j}x_j}}{aii}$$
#### Planteamos un ejemplo de sistema de ecuaciones para comprobar que nuestro algoritmo lo resuelva:
\begin{align}
x+2y+z & =5 \\
y-z & =2 \\
z &=0\\
\end{align}
#### Usaremos el método de sustitución regresiva para resolver sistemas lineales de n ecuaciones con n incógnitas, en el cual la matriz de coeficientes posee ceros debajo de la diagonal 
#### Escribimos en forma matricial:
$$\left(\begin{array}{ccc} a_{11} & a_{12} & a_{13} \\ 0 & a_{22} & a_{23} \\ 0 & 0 & a_{33}\end{array}\right)
\left(\begin{array}{c}x_1\\x_2\\x_3\end{array}\right)=
\left(\begin{array}{c}b_1\\b_2\\b_3\end{array}\right)$$
#### Escribimos en forma aumentada:
$$\left(\begin{array}{ccc} a_{11} & a_{12} & a_{13} &\mbox{|}\ b_1 \\ 0 & a_{22} & a_{23} & \mbox{|}\ b_2 \\
0 & 0 & a_{33} & \mbox{|} \ b_3\end{array}\right)$$
#### Ahora programamos el método:

```python
#Importamos librerias
import numpy as np
#Escribimos la matriz de nuestro sistema A:
A=np.array([[1,2,-2],[0,1,-1],[0,0,1]])
#Ahora se escriben los términos independientes:
b=np.array([5,2,0])
#Para hallar el número de ecuaciones formamos una matriz 3x3:
n=np.size(b)
#Definimos un vector para guardar la solución:
x=np.zeros(n)
#Determinamos el bucle que va desde la última ecua. hasta la primera:
for i in range(n-1,-1,-1):
    sumj=0
    for j in range(i+1,n):
        sumj += A[i,j]*x[j]
    x[i]=(b[i]-sumj)*1/A[i,i] #Planteamos la fórmula
print(x)
```

#### En nuestro ejemplo de igual manera x=1; y=2 y z=0. Por lo tanto, es correcto
#### El costo computacional de un algoritmo está dado por el número de operaciones básicas (llamado flops) 
#### En un sistema con matriz triangular superior de orden nx3 como el que realizamos antes (3x) aplicando el método de sustitución hacia delante el coste computacional es $n^2=3^2= 9$ operaciones. 

### 4. Una importante parte de la física es la termodinámica que es el estudio de la transferencia de calor. En este ejercicio vamos a determinarla distribución de temperatura de estado estable de una placa delgada cuando se conoce la temperatura en los bordes. Suponga que la placa que se ilustra en la figura representa una sección transversal de una viga de metal, con flujo de calor despreciable en la dirección perpendicular a la placa. Sean T1, T2, T3 y T4 las temperaturas en los cuatro nodos interiores de la malla en la figura. La temperatura en un nodo es aproximadamente igual al promedio de las temperaturas de los cuatro nodos más cercanos, esto es, a la izquierda, arriba, a la derecha y abajo. Escriba un sistema de ecuaciones cuya solución de estimaciones de las temperaturas T1, T2, T3 y T4. Programe el método de eliminación gaussiana y halle el valorde estas temperaturas. Utilice comoapoyo el ejercicio del literal anterior.

#### Calculamos el promedio de los nodos más cercanos: 
* $T1=\frac{10+T2+20+T3}{4}$
#### $4T1= 30+T2+T3$ 
#### $4T1-T2-T3=30$ Primera ecuación.
* $T2= \frac{T1+20+40+T4}{4}$
#### $4 T2= 60+T1+T4$
#### $-T1+4T2-T4=60$ Segunda ecuación.
* $T3= \frac{T1+10+T4+30}{4}$
#### $4 T3= 40+T1+T4$
#### $-T1+4T3-T4=40$ Tercera ecuación.
* $T4= \frac{T3+T2+40+30}{4}$
#### $-T2-T3+4T4=70$ Cuarta ecuación.
* Escribimos nuestro sistema de ecuaciones:
\begin{align}
4T1-T2-T3 = 30 \\
-T1+4T2-T4 =60 \\
-T1+4T3-T4 =40\\
-T2-T3+4T4 =70\\
\end{align}
* Escribimos en forma matricial:
$$\left(\begin{array}{cccc} 4 & -1 & -1 & 0 \\ -1 & 4 & 0 & -1 \\ -1 & 0 & 4 & -1 \\ 0 & -1 & -1 & 4\end{array}\right)
\left(\begin{array}{c}T_1\\T_2\\T_3\\T_4\end{array}\right)=
\left(\begin{array}{c}30\\60\\40\\70\end{array}\right)$$
#### Escribimos el algoritmo para la eliminicación gaussiana para hallar el valor de las temperaturas:

```python
#Eliminación Gaussiana
#Función para mostrar la matriz

def mostrarMatriz(matriz,orden):
    for i in range(0,orden):
        linea = "| "
        for j in range(orden+1):
            linea +=  str(matriz[i][j])+" "
        linea += "| "
        print(linea)

matriz = [[4,-1,-1,0,30], [-1,4,0,-1,60], [-1,0,4,-1,40],[0,-1,-1,4,70]]
#Orden de la matriz
orden=len(matriz) 
mostrarMatriz(matriz,orden);    
#Recorrer la matriz 
for j in range(0,orden+1):
    for i in range(0, orden):
        if i>j:
            #Divir los elementos de la matriz
            division=matriz[i][j]/matriz[j][j]
            for k in range(0, orden+1):
                #Obterner el nuevo valor para los elementos en la diagonal inferior
                matriz[i][k]=matriz[i][k]-division*matriz[j][k];
#Recorrer la matriz
x = [0,0,0,0]
for i in range(orden,0,-1):
    suma=0
    for j in range(i,orden):
        suma=suma+matriz[i-1][j]*x[j]
    #Obtener los valores de las variables
    x[i-1]=((matriz[i-1][orden]-suma)/matriz[i-1][i-1])	
#Mostrar los valores de las variables
for i in range(0,orden):
    print("x"+str(i)+" = "+str(x[i]))
```

### 5. Programe el pivoteo parcial en el algoritmo de eliminación gaussiana.Utilice este nuevo programa para resolver el ejercicio anterior.Compare los resultados.


```python
# Método de Gauss-Jordan con pivoteo parcial
#Insertamos librerias
import numpy as np

# Ingresamos los datos de las matrices
A = np.array([[4,-1,-1,0],
              [-1,4,0,-1],
              [-1,0,4,-1],
              [0,-1,-1,4]])

B = np.array([[30],
              [60],
              [40],
              [70]])

# Realizamos el procedimiento
casicero = 1e-15 # Lo totamos como 0

# Evitar truncamiento en operaciones
A = np.array(A,dtype=float) 

# Matriz aumentada
AB = np.concatenate((A,B),axis=1)
AB0 = np.copy(AB)

# Realizamos el pivoteo parcial por filas
tamano = np.shape(AB)
n = tamano[0]
m = tamano[1]

# Para cada fila en AB
for i in range(0,n-1,1):
    # columna desde diagonal i en adelante
    columna = abs(AB[i:,i])
    dondemax = np.argmax(columna)
    
    # dondemax no está en diagonal
    if (dondemax !=0):
        # intercambia filas
        temporal = np.copy(AB[i,:])
        AB[i,:] = AB[dondemax+i,:]
        AB[dondemax+i,:] = temporal
        
AB1 = np.copy(AB)

# eliminacion hacia adelante
for i in range(0,n-1,1):
    pivote = AB[i,i]
    adelante = i + 1
    for k in range(adelante,n,1):
        factor = AB[k,i]/pivote
        AB[k,:] = AB[k,:] - AB[i,:]*factor
AB2 = np.copy(AB)

# elimina hacia atras
ultfila = n-1
ultcolumna = m-1
for i in range(ultfila,0-1,-1):
    pivote = AB[i,i]
    atras = i-1 
    for k in range(atras,0-1,-1):
        factor = AB[k,i]/pivote
        AB[k,:] = AB[k,:] - AB[i,:]*factor
    # diagonal a unos
    AB[i,:] = AB[i,:]/AB[i,i]
X = np.copy(AB[:,ultcolumna])
X = np.transpose([X])


# Imprimimos las salidas
print('Matriz aumentada:')
print(AB0)
print('Pivoteo parcial por filas')
print(AB1)
print('eliminacion hacia adelante')
print(AB2)
print('eliminación hacia atrás')
print(AB)
print('solución de X: ')
print(X)
```

#### El pivoteo parcial dicta que el elemento pivote que debe escogerse es el mayor absolutamente de cada columna. En este ejercicio por casualidad la fila que tiene mayor número ya se encuentra en el primer lugar de la columna, por lo tanto al aplicar el pivoteo parcial por filas la matriz queda de la misma manera que se plantea al inicio. Es así como en este caso el pivoteo parcial mantuvo de igual manera la matriz original. Posteiormente, en el segundo elemento debe también ser el mayor de las columnas para ubicarlo como pivote y volver a realizar la eliminación. Así sucesivamente hasta que no haya elementos bajo el pivote (como observamos en la eliminación hacia adelante) 
#### En conclusión, debido a la disposición de la matriz plateada que ya se encontraba con el pivote primer pivote, los resultados son los mismos mediante ambos métodos (pivoteo parcial y eliminación gaussiana). Es importante señalar que este método tiene menores errores de redondeo o propagación que los que se presentan con el método de eliminación gaussiana, esto se refleja mediante la siguiente propiedad: $|Mik|<=1$, para todo i,tal que $k+1<=i<=n$
